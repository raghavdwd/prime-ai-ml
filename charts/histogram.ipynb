```python
#Histogram
import numpy as np;
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

scores = np.random.normal(70,10,100)
bins = [30,40,50,60,70]
```

```python
plt.hist(scores, color="red", edgecolor="black", bins=bins)
plt.ylabel("# of students")
plt.xlabel("scores")
```

```python
#legit transactions
legit_transactions = [2.88, 3.14, 2.99, 3.22, 3.01, 2.95, 3.05, 300, 298, 310,  3.09, 3.04, 2.94, 3.07, 3.01, 340, 334, 350, 360, 355, 345, 365, 370, 380, 375, 385, 390, 395, 400, 410]

#fraudulent transactions
fraudulent_transactions = [2.85, 3.20, 2.90, 3.25, 3.00, 2.92, 3008, 3.02, 2097, 3012, 3.11]
```

```python
plt.hist(legit_transactions, color="blue", edgecolor="black", bins=10, alpha=0.5, label="Legit Transactions")
plt.hist(fraudulent_transactions, color="red", edgecolor="black", bins=10, alpha=0.5, label="Fraudulent Transactions")
plt.xlabel("Transaction Amount($)")
plt.ylabel("# of Transactions")
plt.legend()
plt.show()
```
