```python
#scatter plot
import matplotlib.pyplot as plt
```

```python
people = ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Heidi', 'Ivan', 'Judy']
ages = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70]
bp = [120, 130, 140, 150, 160, 170, 180, 110, 200, 210]

colors = ['red' if age > 50 else 'blue' for age in ages]

plt.scatter(ages, bp, s=bp, cmap="OrRd", c = bp)
plt.xlabel('Age')
plt.ylabel('Blood Pressure')
plt.title('Age vs Blood Pressure')
plt.grid()
plt.colorbar(label='Blood Pressure')

for i, person in enumerate(people):
    plt.annotate(person, (ages[i], bp[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.ylim(100, 220)
plt.xlim(20, 80)
plt.show()
```
