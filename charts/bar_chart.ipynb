```python
#bar chart
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
```

```python
movies = ['The Shawshank Redemption', 'The Godfather', 'The Dark Knight', 'Pulp Fiction', 'The Lord of the Rings: The Return of the King']
ratings = [9.3, 3.2, 10.0, 5.9, 8.9]
```

```python
with plt.style.context('ggplot'):
    plt.bar(movies, ratings, color='blue')
    plt.xlabel('Movies')
    plt.ylabel('Ratings')
    plt.title('Top 5 Movies and their Ratings')
    plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
    plt.ylim(0, 15)
    plt.tight_layout()
    plt.text(ratings[3], ratings[3] + 10, str(ratings[3]), ha='center')
    plt.show()
```

```python
import numpy as np
```

```python
oscar_movies = ['The Shape of Water', 'Green Book', 'Parasite', '1917', 'Joker']
oscar_revenues = [195.2, 321.8, 258.8, 384.9, 107.4]

non_oscar_movies = ['Avengers: Endgame', 'Highest Rating','The Lion King', 'Frozen II', 'Toy Story 4']
non_oscar_revenues = [2797.5, 1656.9, 1450.0, 1073.4, 1132.5]

year = ['2018', '2019', '2020', '2021', '2022']

width = 0.4

x = np.arange(len(year))

plt.bar(x-width/2, oscar_revenues, width, color='green', label='Oscar-winning Movies')
plt.bar(x + width/2, non_oscar_revenues, width, color='orange', label='Non-Oscar-winning Movies', alpha=0.7)
plt.xlabel('Year')
plt.ylabel('Box Office Revenue (in millions)')
plt.title('Box Office Revenue of Oscar-winning vs Non-Oscar-winning Movies (2018-2022)')
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#horizontal bar chart
plt.barh(movies, ratings, color='purple')
plt.xlabel('Ratings')
plt.ylabel('Movies')
plt.title('Top 5 Movies and their Ratings')
plt.xlim(0, 15)
plt.tight_layout()
plt.text(ratings[3] + 0.5, 3, str(ratings[3]), va='center')
plt.show()
```
