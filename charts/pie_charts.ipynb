```python
import matplotlib.pyplot as plt
```

```python
expenses = [200, 150, 300, 250, 400]
categories = ['Rent', 'Groceries', 'Utilities', 'Entertainment', 'Misc']
```

```python
plt.pie(expenses, labels=categories, autopct='%1.1f%%',wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '--'}, startangle=140,shadow=True, explode=[0.1, 0, 0, 0, 0], colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
plt.title('Monthly Expenses Breakdown')
plt.axis('equal')
plt.show()
```
