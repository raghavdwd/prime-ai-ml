```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

```python
plt.plot(x, y, marker='o')
plt.title("Sample Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

```python
#oscar movies data
oscar_movies = ["Casablanca", "Gone with the Wind", "Lawrence of Arabia", "The Godfather", "The Silence of the Lambs"]
oscar_wins = [3, 10, 7, 3, 5]
oscar_revenue = [1.0, 3.3, 2.5, 2.8, 1.2]  # in billions
oscar_years = [1939, 1942, 1962, 1972, 1991]

plt.figure(figsize=(10, 6))
plt.plot(oscar_movies, oscar_wins, marker='o', label='Oscar Wins')
plt.plot(oscar_movies, oscar_revenue, marker='o', label='Box Office Revenue (Billion $)')
plt.title("Oscar Movies: Wins vs Revenue")
plt.xlabel("Movies")
plt.ylabel("Count / Revenue")
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
#non-oscars movies data
non_oscar_movies = ["Inception", "The Dark Knight", "Pulp Fiction", "The Matrix", "Fight Club"]
non_oscar_revenue = [2.8, 1.0, 0.9, 1.6, 0.3]
years = [1939,1942, 1962, 1972, 1991]

plt.plot(years, non_oscar_revenue, marker='o', label='Non-Oscar Movies')
plt.plot(oscar_years, oscar_revenue, marker='o', label='Oscar Movies')
plt.title("Non-Oscar Movies: Revenue Over Time")
plt.xlabel("Year")
plt.ylabel("Box Office Revenue (Billion $)")
plt.xticks(years)
plt.legend()
plt.grid()
# plt.show()
plt.savefig("oscar_movies_plot.png")
```
