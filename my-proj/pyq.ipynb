```python
import requests
from bs4 import BeautifulSoup

# Configuration variables
LINK_TO_DOWNLOAD = "https://www.pichlequestion.com/ignou/ignou-bcs-041-question-papers-ffaffa8120be11efaa"
OUTPUT_HTML_FILE = "pichlequestion.html"
PAPERS_FOLDER = "BCS-041_Papers"
OUTPUT_PDF_NAME = "Computer_Networks_Combined_PYQs.pdf"
```

```python
try:
    response = requests.get(LINK_TO_DOWNLOAD)
    if response.status_code == 200:
        with open(OUTPUT_HTML_FILE, "w", encoding="utf-8") as file:
            file.write(response.text)
        print("HTML content saved to", OUTPUT_HTML_FILE)
    else:
        print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
except requests.exceptions.RequestException as e:
    print(f"Error fetching the webpage: {e}")
```

```python
try:
    htmlContent = ""
    with open(OUTPUT_HTML_FILE, "r", encoding="utf-8") as file:
        htmlContent = file.read()

    soup = BeautifulSoup(htmlContent, "lxml")
    print(soup.prettify())
except FileNotFoundError:
    print(f"File {OUTPUT_HTML_FILE} not found.")
except Exception as e:
    print(f"Error reading or parsing HTML: {e}")
```

```python
import gdown, time
```

```python
def download_file(url, filename):
    if not url:
        print(f"Invalid URL for {filename}")
        return
    try:
        gdown.download(url, filename, quiet=False)
        time.sleep(2)  # Sleep for 2 seconds to avoid overwhelming the server
        print(f"Downloaded: {filename}")
    except Exception as e:
        print(f"Failed to download {filename}: {e}")
        print(f"URL attempted: {url}")
```

```python
from pypdf import PdfWriter

merger = PdfWriter()
```

```python
def merge_pdfs(pdf_containing_folder, output_filename):
    try:
        if not os.path.exists(pdf_containing_folder):
            print(f"Folder {pdf_containing_folder} does not exist.")
            return
        pdf_files = [f for f in os.listdir(pdf_containing_folder) if f.endswith(".pdf")]
        if not pdf_files:
            print(f"No PDF files found in {pdf_containing_folder}.")
            return
        for filename in pdf_files:
            merger.append(os.path.join(pdf_containing_folder, filename))
        merger.write(output_filename)
        merger.close()
        print(f"Merged PDFs into {output_filename}")
    except Exception as e:
        print(f"Error merging PDFs: {e}")
        merger.close()
```

```python
merge_pdfs(PAPERS_FOLDER, OUTPUT_PDF_NAME)
```
