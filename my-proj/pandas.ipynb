## EDA(EXPLORATORY DATA ANALYSIS)
 - load and structure data and understand
 - clean the data

```python
s = pd.Series([10, 20, 30, 40, 50])
print(s)
```

```python
s = pd.Series([10, 20, 30, 40, 50], index=['adam', 'bie', 'charlie', 'david', 'eve'])
print(s)
```

```python
#vectorized operations
s = pd.Series([10, 20, 30, 40, 50])
print(s * 2) #multiply each element by 2
print(s + 5) #add 5 to each element
print(s / 10) #divide each element by 10
```

```python
s1 = pd.Series([1, 2, 3, 4, 5])
s1[0] = 10
print(s1)
```

```python
changed_s1 = s1.drop(0) #drop the first element
print(changed_s1)
```

```python
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "Los Angeles", "Chicago"]
}
d = pd.DataFrame(data)
print(d)
print (d["Name"]) #access the "Name" column
print (d["Age"]) #access the "Age" column
print (d["City"]) #access the "City" column
print(d, type(d))
print(d.index)
print(d.columns)
print(d.values)
```
