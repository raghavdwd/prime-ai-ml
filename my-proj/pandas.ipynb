## Loc vs iLoc
| Feature         | `loc`              | `iloc`                        |
| --------------- | ------------------ | ----------------------------- |
| Based on        | **Labels (names)** | **Positions (index numbers)** |
| Works with      | Row/column names   | Integer positions             |
| Slice behavior  | **Inclusive**      | **Exclusive (like Python)**   |
| Accepts boolean | ✅ Yes              | ✅ Yes                         |


```python
df_new.loc[0:5, ["city", "country"]] #selecting first 5 rows and specific columns using loc
```

```python
df_new.iloc[0:3,1:5]   #selecting first 3 rows and columns from index 1 to 4 using iloc
```

```python
df_new.at[0, "city"] #selecting a single value using at
```

```python
df_new.iat[0, 2] #selecting a single value using iat where i is integer position
```

```python
cities = df_new["city"].copy() #copying the "city" column to a new variable
print(cities)
cities[0] = "New Delhi" #modifying the first element of the copied series
print(cities) #printing the modified series
print(df_new["city"]) #printing the original "city" column to show it is unchanged
```

```python
#filtering data
df = pd.read_csv('globalAirQuality.csv')
```

```python
df[df["aqi"] > 100] #filtering rows where aqi is greater than 100
```

```python
df[(df["aqi"] > 100) & (df["city"] == "Delhi")] #filtering rows where aqi is greater than 100 and city is Delhi
```

```python
# query method - returns copy of original data
aqi_val = 100
df.query("aqi>100 & temperature>30")[['city', 'country']]
df.query("aqi > @aqi_val & temperature>30")[['city', 'country']]
```
