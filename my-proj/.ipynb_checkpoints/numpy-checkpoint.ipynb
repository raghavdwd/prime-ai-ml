```python
#execution performance
size = 10_000_000

#python list 
py_list = list(range(size))
start = time.time()
sq_list = [x**2 for x in py_list]
end = time.time()
print(f"python list performance time = {end-start} seconds")

#np arrays
np_arr = np.array(py_list)
start = time.time()
sq_array = np_arr ** 2
end = time.time()
print(f"np array performance time = {end-start} seconds")
```
